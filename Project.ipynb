```python
import pandas as pd
import numpy as np
```

```python
# Importing raw data table


rep = pd.read_csv ('Pdata.csv',delimiter = '	')
print(rep)
```

```python
# Top 10 Countries with highest covid cases

grouped = rep.groupby('COUNTRY').sum()
top10 = grouped.sort_values(by='TC',  ascending=False)
top10['TC'][:10]
```

```python
# Countries with least amount of cases

top10['TC'][-11:]
```

```python
# Top 10 conutries with highest covid deaths

top10['TD'][:10]
```

```python
# Countries with least covid deaths

top10['TC'][-11:]
```

```python
# Percentage of total cases among top 10 countries

top10['TC']/sum(top10['TC'])*100
```

```python
# Percentage of total deaths among top 10 countries

top10['TD']/sum(top10['TD'])*100
```

```python
# Death rate due to covid

rep['TD'].diff()
```

```python
# Countries and their max STI

grouped = rep.groupby('COUNTRY').max(['STI'])
grouped
```

```python
# Top 20 countries in terms of STI

grouped = rep.groupby('COUNTRY').max(['STI'])
top20STI = grouped['STI'].sort_values(ascending = False)
top20STI[:20]
```

```python
#Countries with least STI

top20STI[-20:]
```

```python
# Population of each country

group = rep.groupby('COUNTRY').max(['POP'])
group
```

```python
# Top 20 countries in terms of population

group = rep.groupby('COUNTRY').max(['POP'])
top20POP = group['POP'].sort_values(ascending = False)
top20POP[:20]
```

```python
# Countries with lowest population

top20POP[-20:]
```

```python
#Countries and their HDI and population

rep.groupby('COUNTRY').max(['HDI'])
```

```python
# Top 10 countries in terms of HDI

gr = rep.groupby('COUNTRY').max(['HDI'])
top10HDI = gr['HDI'].sort_values(ascending = False)
top10HDI[:10]
```

```python
# Countries with least HDI

top10HDI[-11:]
```

```python
# Countries with their GDP per Capita

gdps = rep.groupby('COUNTRY').max(['gdp_per_capita'])
gdps
```

```python
# Top 20 countries in terms of gdp per capita

gdps['gdp_per_capita'].sort_values(ascending = False)[:20]
```

```python
# Countries with least gdp_per_capita

gdps['gdp_per_capita'].sort_values()[:40]
```

From analysis of the given data, we saw the top countries who encountered the maximum number of covid-19 cases and the maximum number of deaths due to covid-19. 

We also showed the countries with the least covid-19 cases and deaths.

The countries with maximum population, GDP per capita, STI and HDI were also displayed.

STI described the extent to which restrictions were made to curb the spread of the virus. At one point of time several countries had announced 100 STI.

The Human Development Index (HDI) is the measure of life expectancy, income and education of the cictizens of a country. 

The countries with higher value of HDI during the covid-19 phase had relatively lower hit than other countries.

Similarly countries with higher GDP per capita means that the effect of the pandemic was relatively lower than other countries.

Countries with the large populations had to take large measures, as they had more chances of spreading of the virus, the STI values of these countries reached very high.
